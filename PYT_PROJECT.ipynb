# 5. Correlation Between Age and Salary + Visualization

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Salary"], alpha=0.6)
plt.title("Age vs Salary Correlation")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
correlation = df["Age"].corr(df["Salary"])
print("Correlation between Age and Salary:", correlation)
```

5 Correlation Between Age & Salary
A strong positive correlation exists between age and salary, meaning that older employees tend to earn more due to experience.
However, some younger employees earn higher-than-average salaries, likely due to specialized skills or key roles.
 Insight: Salary increments align with experience, but highly skilled younger employees are also compensated well, reflecting a skill-based pay structure.
